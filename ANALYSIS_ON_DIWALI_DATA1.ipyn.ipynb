```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Data provided by the user
data = {
    'Demographic_Combination': [
        'F_Uttar Pradesh_Single_26-35_Healthcare',
        'F_Maharashtra_Single_26-35_Healthcare',
        'F_Uttar Pradesh_Single_26-35_Aviation',
        'F_Uttar Pradesh_Married_26-35_Aviation',
        'F_Maharashtra_Single_26-35_Aviation'
    ],
    'Amount': [
        553427.0,
        493046.0,
        419023.0,
        416406.0,
        386758.0
    ]
}

df_custom_data = pd.DataFrame(data)

# Sort the data in descending order by Amount
df_custom_data = df_custom_data.sort_values(by='Amount', ascending=False)

# Create a colorful and highlighted bar graph
plt.figure(figsize=(14, 8))
sns.barplot(x='Demographic_Combination', y='Amount', data=df_custom_data, palette='viridis', hue='Demographic_Combination', legend=False)

plt.title('Top Spending Combinations by Demographic (Custom Data)', fontsize=16)
plt.xlabel('Demographic Profile', fontsize=12)
plt.ylabel('Total Amount Spent', fontsize=12)
plt.xticks(rotation=45, ha='right', fontsize=10) # Rotate and adjust font size for better readability
plt.yticks(fontsize=10)

# Add annotations on top of each bar to display the amount
for index, row in df_custom_data.reset_index(drop=True).iterrows():
    plt.text(index, row['Amount'], f'{row['Amount']:.2f}', color='black', ha='center', va='bottom', fontsize=10)

plt.tight_layout()
plt.show()
```

### Conclusion: Top Spending Demographic Profiles

Based on the top 5 combined demographic spending profiles:

*   The **highest spending** combination is **F_Uttar Pradesh_Single_26-35_Healthcare**, with a total of **553,427.0**.
*   The **lowest spending** among these top 5 is **F_Maharashtra_Single_26-35_Aviation**, with a total of **386,758.0**.

**Key Insight:** All of the top 5 highest spending combinations are from **Female** customers in the **26-35 age group**. These top spenders are concentrated in **Uttar Pradesh** and **Maharashtra**, and are found in **Healthcare** and **Aviation** occupations, with a mix of **Single** and **Married** marital statuses. This highlights that young adult females, especially in these regions and professions, represent a highly valuable customer segment.
